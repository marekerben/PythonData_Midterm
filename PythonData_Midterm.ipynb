```python
df2 = pd.read_csv('https://ies-python-midterm.s3.eu-central-1.amazonaws.com/companies/companies_no_subindustry.csv')
df2
```

```python
#1
not_included = df2.included.isnull().sum()
print(f'{not_included} companies have no filled in the date of inclusion.')
```

```python
df2 = df2[df2.included.notnull()].copy()
df2
```

```python
df2.included = pd.to_datetime(df2.included, dayfirst=True,errors='coerce')
df2[df2.included == df2.included.min()]
```

```python
df2[df2.included == df2.included.max()]
```

```python
years_included = [2023 - i.date().year for i in df2.included]
add = 0
for i in years_included:
    add = add + i

mean = add / len(years_included)
print(f'{mean} is the average age of constituent in a sample.')
```
